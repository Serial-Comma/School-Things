## 11.4 List Slicing

* As with strings, lists can be sliced to obtain sublists .
* The general principles of slicing in strings also apply to lists.

`[a : b]`  | Obtain sublist from index a to index b-1 (does not include element at index b)

`[a : ]`   | Obtain sublist from index a to end of list.

`[: b]`    | Obtain sublist from start of list to index b-1 (does not include element at index b)

## 11.5 List Operations

### 11.5.1 Length of List using `len()`

* The `len()` method returns the number of elements in a list.

**Example 7**

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j', 'k']
print(len(letters))
len(letters)
```

* The `len()` method can also be used to find the length of a sublist of a nested list

**Example 8**

```python
nums = [[1, 2, 3, 4, 5], ["one", "two", "three", "four", "five"]]
print(len(nums))
print(len(nums[0]))  # nums[0] is the sublist [1, 2, 3, 4, 5]
print(len(nums[1]))  # nums[1] is the sublist ["one", "two", "three", "four", "five"]
```

* In addition, the `len()` method can be used to find the length of an element in a sublist where appropriate. 
* If the element belongs to a data type that has no length attribute, it returns a `TypeError`

**Example 9**

```python
nums = [[1, 2, 3], "one", "two", "three"]
print(f'{nums} has a length of {len(nums)}')
print(f'{nums[0]} has a length of {len(nums[0])}')
print(f'{nums[1]} has a length of {len(nums[1])}')
print(f'{nums[2]} has a length of {len(nums[2])}')
print(f'{nums[3]} has a length of {len(nums[3])}')
print(f'{nums[2][1]} has a length of {len(nums[2][1])}')  # 'w' in 'two' is a string of length 1
print(f'{nums[0][1]} has a length of {len(nums[0][1])}')  # 2 in [1, 2, 3] is an integer. An integer has no length. Returns TypeError.
```

## 11.7 Copy Operations

Note: There may not be a lot of applications for this now yet but it might come in handy when you are dealing with more complication tasks, such as passing the list as an argument to a function etc.

### 11.7.1 Copy by Slicing

* Copying a list by slicing creates a new object in a new memory location. While the duplicated list is the same as the original list, it exists at a different memory address.

**Example 32**

```python
nums_1 = [0, 1, 2, 3]
nums_2 = nums_1[ : ]
print(nums_1 == nums_2)
print(nums_1 is nums_2)
```

### 11.7.2 Copy by Constructor `list()`

* Copying a list using the constructor `list()` also creates a new object in a new memory location. While the duplicated list is the same as the original list, it exists at a different memory address.

**Example 33**

```python
nums_1 = [0, 1, 2, 3]
nums_2 = list(nums_1)
print(nums_1 == nums_2)
print(nums_1 is nums_2)
```

### 11.7.3 Copy using `copy()` - Shallow Copy

* Copying a list using the `copy()` method also creates a new object in a new memory location. While the duplicated list is the same as the original list, it exists at a different memory address.

**Example 37**

```python
nums_1 = [1, 2, 3, 4]
nums_2 = nums_1.copy()
# nums_2[1] = 5
print(nums_1)
print(nums_2)
print(nums_1 == nums_2)
print(nums_1 is nums_2)
```

* Do note that the `copy()` method only performs shallow copy. 
* It copies the list itself, which contains references to the objects in the list. 
* If an object in the list is mutable, a change in the mutable objected will be reflected in both lists.

```python
nested_1 = [1, [2, 3, 4], 5]
nested_2 = nested_1.copy()
print(nested_1)
print(nested_2)
print(nested_1 == nested_2)
print(nested_1 is nested_2)

nested_2[1][1] = 9
print(nested_1)
print(nested_2)
print(nested_1 == nested_2)
print(nested_1 is nested_2)

nested_2.append(6)
print(nested_1)
print(nested_2)
print(nested_1 == nested_2)
print(nested_1 is nested_2)
```
