## 15.3 Modes of File Operation

You can specify the mode used to open a file by applying a second argument to the open function.
* `r`/`w`/`a`: Are you `r`eading, `w`riting or `a`ppending a file?
* `t`/`b`: Is it a `t`ext or `b`inary file?

To do so, we use `open(filepath, mode)` where mode can be `r`/`w`/`a`/`t`/`b`

### 15.3.1 Read / Write / Append

The `mode` specifies how you want to work with the file.
* `'r'`: read mode, which is the default. 
* `'w'`: write mode, for overwriting the contents of a file. **Existing file content will be lost.**
* `'a'`: append mode, for appending new content to the end of the file. Existing content in the file will not be lost.

<u>**File Path**</u>

* If the file is (to be) located in the exact folder where the Python kernel is operating in, only the file name needs to be specified.

* If the file is (to be) located in a sub-folder of the folder where the Python kernel is operating in, specify the portion of the path as that which comes after the folder where the Python kernel is operating in

    <u>**Example**</u>
    * Python kernel is operating in `'C:/Users/Desktop/Practical_21'` 
    * The file `sample.txt` is located in the `Sample_A` sub-folder of the `Practical_15` folder
    * Path to specify is `'/Sample_A/sample.txt'`

  <u>**Example**</u>
    * Python kernel is operating in `'C:/Users/Desktop/Practical_15'` 
    * The file `sample.txt` is located in the `Test` sub-folder within the `Sample_A` sub-folder of the `Practical_15` folder
    * Path to specify is `'/Sample_A/Test/sample.txt'` 


* If the file is (to be) located outside of the folder where the Python kernel is operating in, specify the full path.

    <u>**Example**</u>
    * Python kernel is operating in `'C:/Users/Desktop/Practical_15'` 
    * The file `sample.txt` is located in `'C:/Users/Documents'`
    * Path to specify is `'C:/Users/Documents/sample.txt'`

<u>**What happens if you do not know the file path for which the Python kernel is operating in?**</u>

<u>**Write mode**</u>

* Open the file in write mode


* If the file does not exist, a new file will be created


* **WARNING**: If the file already exists, it will be overwritten (all previous contents gone!)

Let us create the file `sample.txt` in the same folder as the current instance of the Python kernel without using the `%%` magic function.

<u>**Example 5**</u>

````python
output_file = open('sample.txt', 'w')
````

* The use of `write()` to output in a file is similar to the use of `print()` to output in the console.  
* The difference is that `write()` does not add a new line by default whereas `print()` does.
* To add a new line, `\n` needs to be written.
* All data entered is written as text string.

````python
output_file.write('Give a man a program, frustrate him for a day.')
output_file.write('Teach a man to program, frustrate him for a lifetime.')
````

**You must always close the file in order to save and register the changes!** 

Closing the file ensures the following:
* Commit all memory operations to disk (especially for operations involving writing and appending).
* Release storage used temporarily for the file operations.
* Unlock files for other uses. When files are being written or appended, the are locked for editing. 

````python
output_file.close()
````

<u>**Append mode**</u>

* Open the file in append mode.


* If the file does not exist, a new file will be created.


* If the file already exists, new content will be added to the end of the file. Contents will not be overwritten.

## 15.4 Context Manager

* Programming commonly involves usage of resources, e.g. file operations, network and database connections etc. 


* It is important to release these resources after usage because they may be limited in supply.


* **Context manager** is a mechanism in Python for the automatic setup and teardown of resources.


* The **`with`** operator allows an object (function or class) to be implemented as a context manager.


* When implemented as a context manager, `f.close()` will be automatically called when `with` code block exits.


* You can use the `f.closed` attribute to check if a file object is closed.

<u>**Example 10**</u>

````python
f = open('sample.txt')
text = f.read()
print(text)
````

````python
print(f.closed)
````

````python
f.close()
````

````python
print(f.closed)
````

<u>**Example 11**</u>

````python
with open('sample.txt') as f:
    text = f.read()
    print(text)
````

````python
print(f.closed)
````

<u>**Exercise 3**</u>

Complete following file operations using the `with` code block:
* Write `"Alexa, "` to a file `test.txt`. This write operator will overwrite any existing contents in the file.
* Append `"Good morning!\n"` to the file `test.txt`.
* Append `"What time is it?"` to the file `test.txt`.
* Read and print out content from the file `test.txt`.

````python
with open('test.txt', 'w') as f:
    f.write('Alexa, ')
    
with open('test.txt', 'a') as f:
    f.write('Good morning!\n')
    f.write('What time is it?')

with open('test.txt', 'r') as f:
    s = f.read()
    print('Print string:')
    print(s)
    print('Print string representation:')
    print(repr(s))
````

The `repr()` function displays the string representation of what you see in the console i.e. what is/are the string(s) being actually typed to obtain the output that you see if you were to use the print function.

````python
with open('test.txt', 'r') as f:
    print('Without repr:')
    print(f.read(10))
    print(f.read(10))
    print(f.read())
    
print()

with open('test.txt', 'r') as f:
    print('With repr:')
    print(repr(f.read(10)))
    print(repr(f.read(10)))
    print(repr(f.read()))
````

## 15.7 Reading Multiple Lines using `for` loop

````python
with open('quote.txt') as f:
    for i in f:
        print(i)
````

<u>**Question**</u>
* Why above there is an empty line between each line?
* How to remove trailing `\n` from each line?

## 15.8 Writing Multiple Lines

To write a list of strings to a file, the `writelines()` method can be used.

<u>**NOTE**</u> 

No character, e.g. `\n`, will be added or removed.

````python
s = ['Hello', 'World', '\nfrom', '\nSingapore']
with open('test.txt','w') as f:
    f.writelines(s)
````

````python
with open('test.txt') as f:
    content = f.read()
    print(content)
````
