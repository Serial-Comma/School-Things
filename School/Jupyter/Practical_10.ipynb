## 10.4 String Formatting

### 10.4.1 The `format()` Method

* Concatenation using the `+` operator and repetition using the `*` operator can be cumbersome, especially when it involves multiple strings being concatenated with different data types. The whitespace required needs to be carefully accounted for and type conversion using `str()` needs to be performed for each piece of data that is not a string. The process is laborious and prone to errors.


* The `format()` method provides a more streamlined and organised way to simplify the process. This method is able to automatically handle type conversion when used appropriately. Hence the `str()` function need not be repeatedly used.


* The syntax takes the form of `"... {} ... {}...".format(*arg1, *arg2, ...)` or `"... {key1} ... {key2}...".format(**kwarg1, **kwarg2, ...)` or a combination of `*arg` and `**kwarg` e.g. `"... {} ... {key2}...".format(*arg1, **kwarg2, ...)`
    * { } is a placeholder. It is used in place of the data or value to be concatenated. The number of placeholders required depends on the number of pieces of data to be concatenated.
    * `*arg` refers to the identifier or value of the variable to be concatenated; 
    * `*kwargs` refers to a key-value pair. The key is used to refer to the value that is to be concatenated. The key is placed within the placeholder {} in this case so that the value it refers to will be concatenated.  

**Exercise 20**

The following codes demonstrate the use of the `format()` method. Observe how the code is written and how the concatenation occurs.

````python
# Code to print output string "You are studying H2 Computing  9659 for years 2022 and 2023.
# Use of format() method with *args 

subject = 'H2 Computing'
syllabus = 9569
jc1_year = 2022
jc2_year = 2023

print("You are studying {} {} for years {} and {}.".format(subject, syllabus, jc1_year, jc2_year))
````

````python
# Code to print output string "You are studying H2 Computing syllabus 9659 for years 2021 and 2022.
# Use of format() method with **kwargs 
# Note that in this case, the order of the variable identifiers do not matter

print("You are studying {subject} {syllabus} for years {jc1_year} and {jc2_year}." \
      .format(subject = 'H2 Computing', jc1_year = 2021, jc2_year = 2022, syllabus = 9569))
````

* The `*args` and `**kwargs` used in the `format()` method can be indexed starting from 0. By specifiying the index in the placeholder, the string can be concatenated in the way we want it to be, regardless of the sequence assumed by the `*args` or `**kwargs`  

**Exercise 21**

Observe the positioning of the `*args` and run the codes.

````python
subject = 'H2 Computing'
syllabus = 9569
jc1_year = 2021
jc2_year = 2022

print("You are studying {0} {1} for years {2} and {3}.".format(subject, syllabus, jc1_year, jc2_year))
print("You are studying {1} {0} for years {2} and {3}.".format(subject, syllabus, jc1_year, jc2_year))
````

### 10.4.2 The `f-string`

* New string output method introduced in Python 3.6.

* Starts with `f`and followed by the string

* Similar to the `format()` method, uses placeholders `{}`.

* Value in `{}` can be a variable, an object or an expression.

**Exercise 22**

Run the following code containing both an f-string output and another created using the `format()` method.

````python
x = 1

# Using f-string
area_1 = f'Circle with radius {x} unit has area {x * x * 3.14} square units.'
print(area_1)

# Using format() method
area_2 = 'Circle with radius {} unit has area {} square units.'.format(x, x * x * 3.14)
print(area_2)
````

* Notice that variables are used inside the placeholder instead of at the end of the string when using the f-string method.

**Exercise 23**

* The `f-string` method works well with classes, modules, functions etc. because of its ablity to handle expressions.

````python
# This is a function that takes in 
def addition(x, y):
    return x + y

x = 4
y = 8
print(f'{x} + {y} = {addition(x,y)}')
````

````python
# The equivalent string using str.format()
print('{} + {} = {}'.format(x, y, addition(x,y)))
````

````python
# Example - Rounding off to specified decimal places.

x = 3.14159265359

# Using f-string
x_round_1 = f'{x:.3f}'
print(x_round_1)

# Using format() method
x_round_2 = '{:.3f}'.format(x) 
print(x_round_2)
````

In this case the `.3` represents the precision for the fixed-point `f` type, which in this case means that there are 3 digits after the decimal point.

**Percentage**

**Exercise 26**

````python
# Example - Converting to percentages

x = 0.45

# Using f-string
x_to_percent_1 = f'{x:%}'
print(x_to_percent_1)

# Using format() method
x_to_percent_2 = '{:%}'.format(x) 
print(x_to_percent_2)

# Precision of 2 d.p.
x_to_percent_3 = f'{x:.2%}'
print(x_to_percent_3)
````

In this case, using `%` multiplies the number `x` by 100 and displays in fixed (`'f'`) format, followed by a percent sign. Note that by default, the fixed `f` format uses a precision of 6 digits after the decimal point for `float`.
